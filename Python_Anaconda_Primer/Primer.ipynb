### Standard Library Stuff

```python
# An object is, loosely speaking, anything that has access to attributes(data) and/or methods(functions) by virtue of
# the kind of thing that it is (it's class).  Python is a strictly object oriented language, which means nothing
# can exist in a Python program that isn't an object; Modules, functions and even simple data types, such as integers,
# are all objects in Python.  We can access attributes/methods by typing a '.' after an objects name (In general X.Y can
# be read as 'Y, which belongs to X' in Python).  Since we have Intellisense, typing an object name, follwed by a '.'
# and 'tab' can tell you alot about what a variable might be or how it's used.

# Moving information between different structures can be thought of as giving existing objects
# new relevant attributes/methods, or removing old irrelevant attributes/methods.
```

```python
# Before we touch on the Anaconda environment, a note about comprehensions seems warranted:
# Python offers a syntax designed to simplify common iterated operations (loops) that reduces character count
# and improves readability.  Using comprehensions can save you weeks of time and reams of non-value
# added code in the long run.  In general, comprehensions follow the structure:

# [thing for thing in iterable_containing_things]
```

```python
# List comprehension example task: Create a list of positive integers less than 10.

numbers_comprehension = [n for n in range(0, 10)]

# is syntactically equivalent to:

numbers_loop = []

for n in range(0, 10):
    
    numbers_loop.append(n)
    
print(numbers_comprehension, numbers_loop) # Should be the same thing
```

```python
# Comprehensions also support function calls and logical operators:

# Example task: Create a list of the square of all odd numbers less than 10.

def square(x): return x**2
def isOdd(x): return bool(x%2)

squares_comprehension = [square(n) for n in range(0, 10) if isOdd(n)]

# is syntactically equivalent to:

squares_loop = []

for n in range(0, 10):
    
    if n%2 != 0:
        
        squares_loop.append(n**2)

print(squares_comprehension, squares_loop)
```

```python
# This same style is also available for dictionary objects: 

odd_numbers_comp = {n: isOdd(n) for n in range(0, 10)}

# is syntactically equivalent to:

odd_numbers_loop = {}

for n in range(0, 10):
    
    odd_numbers_loop[n] = n%2 != 0

print(odd_numbers_comp, '\n', odd_numbers_loop)
```

### NetworkX - Graph Theory and Routing
### Documentation: https://networkx.github.io/documentation/stable/

```python
# I think Graph Theory has a very expressive way of describing classical logistics problems; we can model
# the behavior of a profit maximizing  ltl planner using locations as nodes, and the cost of transportation
# as the weighted edges between those nodes.
```

```python
import networkx as nx
import random # library for making random numbers
import itertools # library for calculating permutations/combinations of sets
```

```python
# Let's make up some fake carrier rate data

ltl_carriers = ['Yellow', 'Old Dominion', 'Roadway', 'Estes']

crossdocks = {

    "Saint Louis": (38.6270, 90.1994),
    "Chicago": (41.8781, 87.6298),
    "New York": (40.7128, 74.0060),
    "Los Angeles": (34.0522, 118.2437),
    "Miami": (25.7617, 80.1918),
    "Atlanta": (33.7490, 84.3880),
    "Seattle": (47.6062, 122.3321),
    "Denver": (39.7392, 104.9903),

}

# Get all non-trivial (Saint Louis -> Saint Louis doesn't mean anything, etc.) combinations of lanes and carriers:
lanes = [i for i in itertools.product(crossdocks, crossdocks, ltl_carriers) if i[0] != i[1]]

# Sample a few lanes at random to avoid a cluttered graph:
lanes = random.sample(lanes, 50)

# Assign rates randomnly (between $ 50 & $ 250, independent of distance) and store everything in a dataframe
rate_table = pd.DataFrame({'Origin': [l[0] for l in lanes], 'Destination': [l[1] for l in lanes],
                           'Carrier': [l[2] for l in lanes], 'Rate': [random.randint(50, 250) for l in lanes]})

rate_table[::5] # Display a few records from our rate table
```

```python
# Now that we have a simple ltl rate table like we might see as a 3PL, we can write the information
# to a graph object:

carrier_network = nx.MultiDiGraph() # initialize an empty graph

for index, row  in rate_table.iterrows(): # Iterate over every row in our table
    
    # Write an edge from the rows Origin to its Destination, with weight equal to cost and accompanying carrier detail
    carrier_network.add_edge(row['Origin'], row['Destination'], weight=row['Rate'], carrier=row['Carrier'])
```

```python
# Let's take a look at what we've made:

pos = {c: (-crossdocks[c][1], crossdocks[c][0]) for c in crossdocks} # convert crossdock locations to useable coordinates

nx.draw_networkx(carrier_network, pos, node_size=50, node_color='Blue', width=0.05) # Visual Representation
```

```python
# We can use Dikjstra's algorithm to find the minimum spanning tree of the graph we've drawn.
# Since the weights on the edges represents the cost to travel the edge, the minimum spanning tree
# corresponds to the lowest cost solution by definition.  While this implementation can't tell you which carrier
# was used to find the cheap solution, we could add that functionality if we desired. 

origin, dest = 'Denver', 'New York' # Set an origin and destination within the network

nx.has_path(carrier_network, origin, dest) # 

shortest_path = nx.dijkstra_path(carrier_network, origin, dest)
cost = nx.dijkstra_path_length(carrier_network, origin, dest)

print(shortest_path, '$'+str(cost)) # Route may not make intuitive sense due to random prices
```

```python
# Additionally we can write constraints into the solution process: if we had corresponding transit requirements
# and quoted carrier times, we could calculate the shortest path offering on-time delivery.  We can also write
# lane specific information to edges like, 'Refrigerated' or 'Pup Truck' and only consider paths that match specific
# customer demands.
```

### Numpy - Working with Vectors/Matricies
### Documentation - https://docs.scipy.org/doc/numpy-1.15.0/

```python
# Numpy is the most common Python implementation for working with non-structured multi-dimensional data.
# Alot of the nasty behind-the-scenes computer science issues being handled by our Pyhton packages are hashed out in
# numpy arrays; being familiar with the common array operations can help you troubleshoot programming in high memory
# overhead situations, but it's not necessary if your primary interest is in writing analytical scripts.

# Note: Many existing machine learning libraries/stat packages will only accept numpy arrays as inputs

import numpy as np
```

```python
# Numpy is also a relatively straightforward way to approach linear algebra problems:

# Example: Solve the following system of equations

# 7*x + 12y -3z = 10
# 10x - 6y + z = -5
# 4z = 8

variables = ['x', 'y', 'z'] # assign names for reference

A = np.array([[7, 10, 0], [12, -6, 0], [-3, 1, 4]]) # Coefficient matrix
B = np.array([10, -5, 8]) # Dependent matrix

solutions = np.linalg.solve(A, B) # lapack routine (no idea how this works under the hood)
solutions = {var: value for var, value in zip(variables, solutions)} # Tie variable names to solutions by index

print(solutions)
```
